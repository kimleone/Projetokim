#Análise exploratória

```
import pandas as pd
url = 'https://raw.githubusercontent.com/kimleone/Projetokim/master/Social_Network_Ads.csv'
```

```
da = pd.read_csv(url)
```

```
da.isnull().mean()
```

```
da.drop(['User ID'], axis = 1, inplace = True)
da.head()
```

```
da.describe()
```

```
da.corr()
```

```
import seaborn as sns
import matplotlib.pyplot as plt
plt.style.use('ggplot')

plt.figure(figsize=(12,7))


plt.subplot(2,2,1)
plt.hist(da[ da['Purchased'] == 0]['EstimatedSalary'], alpha = .75)
plt.hist(da[ da['Purchased'] == 1]['EstimatedSalary'], alpha = .75)
plt.title('Histogram of estimated salary according to purchased.')


plt.subplot(2,2,2)
sns.boxplot( x = 'Purchased', y = 'Age', data = da)
plt.title('Boxplot of age')

plt.subplot(2,2,3)
sns.barplot(x='Purchased', y = 'Gender', data = da)
plt.title('Comparação dos Sexos em relação à Compras', size = 15)


plt.subplot(2,2,4)
plt.scatter( x = da['Age'], y = da['EstimatedSalary'], c = da['Purchased'],
            marker = 'o', alpha = 0.5, cmap = 'rainbow' )
plt.title('Dispersion of age and estimated salary.')

plt.tight_layout()
plt.show()
```
